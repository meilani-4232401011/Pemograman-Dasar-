```python
# soal no 2 (hitung_efisiensi)
def hitung_efisiensi(input_energi, output_energi):
    efisiensi = output_energi / input_energi
    return efisiensi

hasil = hitung_efisiensi(500, 400)
print("Efisiensi turbin =", hasil)
```

```python
# Soal no 6
# Program menghitung daya listrik (P = V x I)

# Input nilai tegangan dan arus
tegangan = 220   # Volt
arus = 15        # Ampere

# Menghitung daya listrik
daya = tegangan * arus

# Menampilkan hasil perhitungan
print("Tegangan (V) =", tegangan, "Volt")
print("Arus (I) =", arus, "Ampere")
print("Daya listrik (P) =", daya, "Watt")
```

Program ini berfungsi untuk menghitung daya listrik menggunakan rumus P=V√óIP = V \times IP=V√óI, di mana P adalah daya (Watt), V tegangan (Volt), dan I arus (Ampere). Nilai tegangan ditetapkan 220 Volt dan arus 15 Ampere. Program kemudian menghitung daya dengan mengalikan keduanya dan menyimpannya dalam variabel daya. Hasil perhitungan, yaitu 3300 Watt, kemudian ditampilkan ke layar menggunakan perintah print(). Program ini menunjukkan penerapan sederhana rumus fisika dalam pemrograman Python.

```python
# soal no 8
# Program menghitung efisiensi termal turbin gas
# Rumus: Efisiensi (%) = (energi keluar / energi masuk) √ó 100

def hitung_efisiensi_termal():
    # Meminta input dari pengguna
    energi_masuk = float(input("Masukkan energi masuk (dalam Joule): "))
    energi_keluar = float(input("Masukkan energi keluar (dalam Joule): "))

    # Menghitung efisiensi
    efisiensi = (energi_keluar / energi_masuk) * 100

    # Menampilkan hasil
    print("\n=== HASIL PERHITUNGAN ===")
    print(f"Energi masuk   : {energi_masuk} Joule")
    print(f"Energi keluar  : {energi_keluar} Joule")
    print(f"Efisiensi termal: {efisiensi:.2f}%")

# Memanggil fungsi
hitung_efisiensi_termal()
```

Program ini digunakan untuk menghitung efisiensi termal turbin gas.
Pada program tersebut, pengguna diminta memasukkan nilai energi masuk dan energi keluar dalam satuan Joule. Misalnya, jika energi masuk bernilai 10.000 Joule dan energi keluar 5.000 Joule. Artinya, turbin gas tersebut bekerja dengan efisiensi sebesar 50%, yang berarti hanya separuh energi yang masuk dapat dimanfaatkan sebagai energi keluaran. Setelah perhitungan selesai, program menampilkan hasil lengkap berupa energi masuk, energi keluar, dan nilai efisiensi termal dengan dua angka di belakang koma. Program ini menunjukkan bagaimana konsep efisiensi energi dapat diterapkan dalam bentuk perhitungan sederhana menggunakan Python.

```python
# soal nomor 10
tangki = 0
kapasitas_maks = 100
pengisian_per_iterasi = 10

while tangki < kapasitas_maks:
    tangki += pengisian_per_iterasi
    if tangki == 80:
        print(f"Volume tangki saat ini: {tangki} liter")
        print("WARNING: Hampir penuh!")
    elif tangki == 100:
        print(f"Volume tangki saat ini: {tangki} liter")
        print("TANGKI PENUH ‚Äì Pengisian berhenti.")
    else:
        print(f"Volume tangki saat ini: {tangki} liter")
```

Penjelasan kode nomor 10 ini menggambarkan simulasi proses pengisian tangki bahan bakar menggunakan perulangan while.

Inisialisasi variabel:

tangki = 0 ‚Üí Menandakan tangki mulai dari kondisi kosong (0 liter).

kapasitas_maks = 100 ‚Üí Menentukan kapasitas maksimum tangki, yaitu 100 liter.

pengisian_per_iterasi = 10 ‚Üí Menunjukkan bahwa setiap proses pengisian akan menambah 10 liter bahan bakar ke dalam tangki.

Logika perulangan:
Perintah while tangki < kapasitas_maks: membuat proses pengisian terus berlangsung selama volume tangki belum mencapai kapasitas maksimum.

Proses pengisian dan pengecekan:
Setiap kali loop berjalan, tangki += pengisian_per_iterasi menambah 10 liter bahan bakar ke tangki. Setelah itu, program memeriksa kondisi berikut:

Jika tangki == 80, maka akan muncul pesan "WARNING: Hampir penuh!" sebagai peringatan.

Jika tangki == 100, program mencetak "TANGKI PENUH ‚Äì Pengisian berhenti.", menandakan tangki telah mencapai kapasitas maksimal, sehingga loop berhenti secara otomatis.

Untuk kondisi selain itu, program hanya menampilkan volume bahan bakar saat ini.

```python
# SOAL CHALLENGE
# Simulasi Biaya Listrik Rumah Tangga
import random

def hitung_biaya_listrik(energi_kwh, tarif_per_kwh):
    """
    Menghitung biaya listrik harian berdasarkan energi yang digunakan.
    """
    return energi_kwh * tarif_per_kwh


def status_penggunaan(energi_harian):
    """
    Memberikan status penggunaan energi harian.
    """
    if energi_harian < 5:
        return "Rendah"
    elif energi_harian <= 10:
        return "Normal"
    else:
        return "Tinggi"


def simulasi_penggunaan_bulanan_otomatis(tarif_per_kwh):
    """
    Mensimulasikan penggunaan listrik otomatis selama 30 hari.
    Energi harian diacak menggunakan random.uniform().
    """
    hari = 1
    total_biaya = 0
    total_energi = 0

    while hari <= 30:
        # Energi acak tiap hari antara 3‚Äì15 kWh
        energi_harian = round(random.uniform(3, 15), 2)
        biaya_harian = hitung_biaya_listrik(energi_harian, tarif_per_kwh)
        status = status_penggunaan(energi_harian)

        print(f"Hari ke-{hari}: {energi_harian:.2f} kWh | Status: {status} | Biaya: Rp{biaya_harian:.2f}")

        if energi_harian > 12:
            print("‚ö†Ô∏è  PERINGATAN: Konsumsi listrik tinggi hari ini!")

        total_energi += energi_harian
        total_biaya += biaya_harian
        hari += 1

    return total_energi, total_biaya


def tampilkan_ringkasan(energi_bulanan, biaya_bulanan):
    """
    Menampilkan ringkasan hasil simulasi bulanan.
    """
    rata_energi = energi_bulanan / 30
    print("\n=== RINGKASAN BULANAN ===")
    print(f"Total Energi: {energi_bulanan:.2f} kWh")
    print(f"Total Biaya: Rp{biaya_bulanan:.2f}")
    print(f"Rata-rata konsumsi per hari: {rata_energi:.2f} kWh")

    if rata_energi < 6:
        print("Kategori: Hemat Energi üëç")
    elif rata_energi <= 10:
        print("Kategori: Normal ‚öôÔ∏è")
    else:
        print("Kategori: Boros ‚ö†Ô∏è")


# ===== PROGRAM UTAMA =====
print("=== SIMULASI BIAYA LISTRIK RUMAH TANGGA ===")

tarif = float(input("Masukkan tarif listrik per kWh (Rp): "))

energi_total, biaya_total = simulasi_penggunaan_bulanan_otomatis(tarif)
tampilkan_ringkasan(energi_total, biaya_total)
```

PENJELASAN SOAL CHALLENGE

Program Simulasi Biaya Listrik Rumah Tangga ini dirancang untuk memperkirakan pemakaian energi listrik harian selama satu bulan (30 hari) secara otomatis, kemudian menghitung total energi dan biaya listrik yang dikeluarkan. Selain itu, program juga memberikan informasi mengenai tingkat penggunaan energi ‚Äî apakah termasuk hemat, normal, atau boros.

1.) Pengacakan Konsumsi Energi Harian (random.uniform(3, 15))

Pada bagian ini, program menggunakan fungsi random.uniform(3, 15) untuk menghasilkan angka acak antara 3 hingga 15 kWh setiap harinya.
Nilai acak ini menggambarkan variasi penggunaan listrik sehari-hari pada rumah tangga, karena pada kenyataannya konsumsi listrik tidak selalu sama tiap hari. Misalnya, pada hari tertentu penggunaan listrik meningkat karena pemakaian AC atau alat elektronik lainnya, sementara di hari lain bisa lebih rendah.
Dengan demikian, penggunaan fungsi acak ini membantu membuat simulasi terasa lebih realistis dan dinamis dibandingkan jika nilai energi dimasukkan secara manual.

2.) Penggunaan Perulangan while

Struktur while loop digunakan untuk menjalankan simulasi selama 30 hari berturut-turut.
Kondisi while hari <= 30: memastikan bahwa proses penghitungan akan berulang dari hari pertama hingga hari ke-30.
Di dalam setiap iterasi perulangan ini, program:

1. Menghasilkan nilai energi harian (dari fungsi acak tadi),
2. Menghitung biaya listrik berdasarkan tarif per kWh yang diinput pengguna,
3. Menampilkan hasil konsumsi dan status penggunaan energi,
4. Serta menambah total energi dan biaya ke dalam variabel penjumlahan bulanan.

Dengan cara ini, perulangan while berperan sebagai inti dari simulasi waktu (30 hari), yang secara otomatis memproses seluruh hari tanpa perlu input manual dari pengguna.
