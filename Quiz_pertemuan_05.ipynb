```python
# soal 5
print("--- Program Menghitung Jumlah Deret Kuadrat ---")

ulangan = 1
batas_ulangan = 3

while ulangan <= batas_ulangan:
    print("\nLatihan ke-", ulangan)

    total = 0
    for angka in range(1, 11):
        total += angka ** 2
        print("Angka:", angka, "→ Kuadrat:", angka ** 2, "→ Total sementara:", total)

    print("Jumlah total kuadrat dari 1 sampai 10 =", total)
    ulangan += 1

print("\nSelesai! Semua latihan telah dilakukan sebanyak 3 kali.")
```
